

## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
def lanzamineto_dados(int): 
   
    lista_puntaje = []

    for i in range(int):
    
        dado_aux_1= random.randint(1,6)
        dado_aux_2= random.randint(1,6)
        puntaje = dado_aux_1 + dado_aux_2
        lista_puntaje.append(puntaje)

    return lista_puntaje



     
```

```python
print(lanzamineto_dados(20))
```

```python
# Solución 1 
print(f"El resultado de lanzamientos es: {sum(lanzamineto_dados(1000))}")
```

```python
# Solución 2 
# (expression for variable in iterable if condition)
print(sum(1 for num in lanzamineto_dados(1000) if num > 7 or num % 2 == 0))
```

```python
#Solucion 3 
resultado = (sum(1 for num in lanzamineto_dados(1000) if num > 7 or num % 2 == 0))

print(resultado/1000)
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python

def maquina_saca_bolas():
    ball_box = []

    for i in range(60):
        if i < 10:
            ball_box.append("White")
        elif (i > 9) and (i < 30):
            ball_box.append("Red")
        else:
            ball_box.append("Green")

    return ball_box
```

```python

probabilidad_1 = 0
probabilidad_2 = 0
intentos = 10000

caja_bolas = maquina_saca_bolas()
    
for i in range(intentos) :   
    Colores_que_salen = [random.choice(caja_bolas) for _ in range(5)]

    bolas_blancas = Colores_que_salen.count("White")
    bolas_rojas = Colores_que_salen.count("Red")
    bolas_verdes = Colores_que_salen.count("Green")

    
    probabilidad_1 += 1 if bolas_blancas == 3 and bolas_rojas == 2 and bolas_verdes == 0 else 0
    probabilidad_2 += 1 if bolas_blancas == 5 or bolas_rojas == 5 else 0


resultado_1 = probabilidad_1 / intentos
resultado_2 = probabilidad_2 / intentos

```

```python
print(f"Probabilidad de sacar 3 blancas y 2 rojas: {resultado_1}")
print(f"Probabilidad de sacar todas del mismo color: {resultado_2}")
```
